### Single inheritance


```python

# Example 1

# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("The animal makes a sound.")

# Child class
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name) 

    def speak(self):
        print(f"The dog named {self.name} barks.")

# Object

my_dog = Dog("Tommy")
my_dog.speak()


print("--"*10)

# Example 2

#Parent Class
class Car:
    def __init__(self, car_details):
        self.brand = car_details.get('brand')
        self.type = car_details.get('type')
        self.top_speed = car_details.get('top_speed')

    def car_run(self):
        print(f'car {self.brand} is {self.type} type and it run at {self.top_speed} top speed maximum')

# Child Class

class RacingCar(Car):
    def __init__(self, car_details):
        super().__init__(car_details)
        


car_details = {
    'brand':'maruti',
    'type' : 'SUV',
    'top_speed':'250 kmph'
}
racing_car_details = {
    'brand':'ferrari',
    'type' : 'sports',
    'top_speed':'350 kmph'
}

# Object
car_info = Car(car_details)
car_info.car_run()  

racing_car_info = RacingCar(racing_car_details)
racing_car_info.car_run()


print("--"*10)


# Exampla 3

# Parent Class
class Employee:
    def __init__(self, name, position):
        self.name = name
        self.position = position

    def work(self):
        print(f"{self.name} is working as a {self.position}.")  


# Child Class
class Manager(Employee):
    def __init__(self, name, position, team_size):
        super().__init__(name, position)
        self.team_size = team_size

    def work(self):
        print(f"{self.name} is managing a team of {self.team_size} members.")  

emp = Employee("Raghaqv", "Developer")

mgr = Manager("Sumit", "Manager", 5)

# Object
emp.work()     
mgr.work()

print("--"*10)

# Example 4

# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Child Class
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def introduce(self):
        print(f"Hello, my name is {self.name}, I am {self.age} years old, and my student ID is {self.student_id}.")     


person = Person("Ajay", 21)
student = Student("Rahul", 25, "2025ARC009")


# Object
person.introduce()
student.introduce() 
```

### Multiple Inheritance

```python
# Example 1

#parent 1
class Writer:
    def __init__(self, language):
        self.language = language

#parent 2
class Doctor:
    def __init__(self, specialization):
        self.specialization = specialization

# Child Class
class Doctor_and_Writer(Doctor, Writer):
    def __init__(self, specialization, language):
        Doctor.__init__(self, specialization)
        Writer.__init__(self, language)

    def info(self):
        print(f"Doctor specialization: {self.specialization} also Writes in: {self.language}")

# Object
doctor_writer = Doctor_and_Writer("Cardiologist", "English")
doctor_writer.info()

print("--"*10)

# Example 2

# Parent 1
class Farmer:
    def __init__(self, crop):
        self.crop = crop

# Parent 2
class Shopkeeper:
    def __init__(self, shop_type):
        self.shop_type = shop_type

# Child Class
class FarmerShopkeeper(Farmer, Shopkeeper):
    def __init__(self, crop, shop_type):
        Farmer.__init__(self, crop)
        Shopkeeper.__init__(self, shop_type)

    def details(self):
        print(f"Farmer grows: {self.crop} and also runs: {self.shop_type} shop")

# Object
farmer_shopkeeper = FarmerShopkeeper("Wheat", "Grocery")
farmer_shopkeeper.details()

print("--"*10)


# Example 3

# Parent 1
class Mother:
    def __init__(self, children_count):
        self.children_count = children_count

# Parent 2
class Indian:
    def __init__(self, state):
        self.state = state

# Child Class
class IndianMother(Mother, Indian):
    def __init__(self, children_count, state):
        Mother.__init__(self, children_count)
        Indian.__init__(self, state)

    def introduce(self):
        print(f"She is an Indian mother from {self.state} with {self.children_count} children.")

# Object
indian_mother = IndianMother(3, "Maharashtra")
indian_mother.introduce()

print("--"*10)

# Example 4

# Parent 1
class Teacher:
    def __init__(self, subject):
        self.subject = subject

# Parent 2
class PTTeacher:
    def __init__(self, sport):
        self.sport = sport

# Child Class
class SportsTeacher(Teacher, PTTeacher):
    def __init__(self, subject, sport):
        Teacher.__init__(self, subject)
        PTTeacher.__init__(self, sport)

    def show(self):
        print(f"Teacher Teaches {self.subject} and also coaches {self.sport}")
    

# Object
pt_teacher = SportsTeacher("Math", "Football")
pt_teacher.show()
```

### Hierarchy  Inheritance

```python
# Example 1

# Parent Class
class Animal:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(f"{self.name} moves.")

# Child 1
class Bird(Animal):
    def fly(self):
        print(f"{self.name} can fly.")

# Child 2
class Fish(Animal):
    def swim(self):
        print(f"{self.name} can swim.")

parrot = Bird("Parrot")
shark = Fish("Shark")

# Object
parrot.move()
parrot.fly()
shark.move()
shark.swim()

print("--"*10)

# Example 2

# Parent Class
class Device:
    def __init__(self, brand):
        self.brand = brand

    def power_on(self):
        print(f"{self.brand} device is powered on.")

# Child 1
class Laptop(Device):
    def code(self):
        print(f"Coding on {self.brand} laptop.")

# Child 2
class Smartphone(Device):
    def call(self):
        print(f"Calling from {self.brand} smartphone.")

dell_laptop = Laptop("Dell")
samsung_phone = Smartphone("Samsung")

# Object
dell_laptop.power_on()
dell_laptop.code()
samsung_phone.power_on()
samsung_phone.call()

print("--"*10)

# Example 3

# Parent Class
class EmployeeBase:
    def __init__(self, name):
        self.name = name

    def attend(self):
        print(f"{self.name} is attending office.")

# Child 1
class Developer(EmployeeBase):
    def develop(self):
        print(f"{self.name} is writing code.")

# Child 2
class Designer(EmployeeBase):
    def design(self):
        print(f"{self.name} is creating designs.")

dev = Developer("Shaam")
des = Designer("Shubham")

# Object
dev.attend()
dev.develop()
des.attend()
des.design()

print("--"*10)


# Example 4

# Parent Class
class Vehicle:
    def __init__(self, model):
        self.model = model

    def start(self):
        print(f"{self.model} vehicle started.")

# Child 1
class Car(Vehicle):
    def drive(self):
        print(f"Driving the car: {self.model}")

# Child 2
class Bike(Vehicle):
    def ride(self):
        print(f"Riding the bike: {self.model}")

car = Car("Toyota")
bike = Bike("Yamaha")

# Object
car.start()
car.drive()
bike.start()
bike.ride()
```
